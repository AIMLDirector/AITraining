```python
from nltk.tokenize import word_tokenize, wordpunct_tokenize, sent_tokenize
```

```python
text1="Neural network applications in medicine, science, and business address problems in pattern classification, prediction, and control and optimization."
```

```python
word_tokenize(text1)
```

```python
wordpunct_tokenize(text1)
```

```python
sent_tokenize(text1)
```

```python
sent_tokenize(text2)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words("english"))      
```

```python
print(stop_words)
```

```python
stop_words1 = set(stopwords.words("french"))      
```

```python
print(stop_words1)
```

```python
stop_words3 = set(stopwords.words("russian"))   
```

```python
print(stop_words3)
```

```python
token_word = word_tokenize(text1)
```

```python
print(token_word)
```

```python
Filtered_list = []
```

```python
for word in token_word:
    if word.casefold() not in stop_words:
        Filtered_list.append(word)
        
        
```

```python
print(Filtered_list)
```

```python
# one liner python coding 
Filtered_list1 = [word for word in token_word if word.casefold() not in stop_words]
print(Filtered_list1)

```

```python
stop_word_test = { '.', ',' }
```

```python
filtered_list_special = []
```

```python
for i in Filtered_list:
    if i not in stop_word_test:
        filtered_list_special.append(i)
```

```python
print(filtered_list_special)
```
